```python
# Ler o arquivo CSV usando pandas
# Como o delimitador é uma tabulação, usamos '\t' como separador
df_dados = pd.read_csv("../data/dropout_data.csv", delimiter='\t')

# Exibir as primeiras linhas do dataframe para verificar se os dados foram carregados corretamente
df_dados.head(1000)
```

```python
# Contagem dos valores nulos
print("Contagem dos valores nulos de cada coluna")
print(df_dados.isna().sum())
```

```python
df_dados['EVA'] = df_dados['EVA'].fillna(0) # Substituir todos os valores NaN na coluna EVA por 0
df_dados['EVA'] = df_dados['EVA'].astype(int) # Converter a coluna EVA para inteiro
```

```python
df_dados['RF'] = df_dados['RF'].str.replace(',', '.') # Substituir , por . no campo RF
df_dados['RF'] = pd.to_numeric(df_dados['RF'], errors='coerce') # Substituir valores não numéricos ('S/I') por NaN
df_dados['RF'] = df_dados['RF'].astype(float) # Converter a coluna RF para FLOAT
```

```python
df_dados['IDA-IN'] = df_dados['IDA-IN'].replace(['S/I', 'ND'], np.nan) # Substituir valores específicos ('S/I' e 'ND') por NaN
df_dados['IDA-IN'] = df_dados['IDA-IN'].replace(['', ' '], np.nan) # Substituir valores específicos ('' e ' ') por NaN
df_dados['IDA-IN'] = pd.to_numeric(df_dados['IDA-IN'], errors='coerce') # Substituir valores não numéricos ('S/I') por NaN
df_dados['IDA-SUP'] = pd.to_numeric(df_dados['IDA-SUP'], errors='coerce') # Substituir valores não numéricos ('S/I') por NaN
df_dados['IDA-MÉD'] = pd.to_numeric(df_dados['IDA-MÉD'], errors='coerce') # Substituir valores não numéricos ('S/I') por NaN
```

```python
df_dados['TE'] = pd.to_numeric(df_dados['TE'], errors='coerce')
df_dados['TE'] = df_dados['EVA'] / df_dados['MAT']
```

```python
df_dados['SEXO'] = df_dados['SEXO'].replace(['S/I', 'ND'], np.nan) # Substituir valores específicos ('S/I' e 'ND') por NaN
df_dados['SEXO'] = df_dados['SEXO'].replace([' ', ''], np.nan) # Substituir valores específicos ('S/I' e 'ND') por NaN
```

```python
df_dados['TO'] = df_dados['TO'].fillna('NSA')
df_dados['RF-CAT'] = df_dados['RF-CAT'].fillna('ND')
```

```python
# Contagem dos valores nulos
print("Contagem dos valores nulos de cada coluna")
print(df_dados.isna().sum())
```

```python
# Lista para armazenar os dados expandidos
dados_expandidos = []

for index, row in df_dados.iterrows():
    # Para cada exemplo na coluna COM (alunos que não evadiram)
    for _ in range(row['COM']):
        # dados_expandidos.append({'AE': 'NÃO', 'COM': row['COM'], 'EVA': row['EVA']})
        nova_linha = row.copy()  # Copiar todos os dados da linha atual
        nova_linha['AE'] = 'NÃO'  # Adicionar a coluna AE com valor "NÃO"
        dados_expandidos.append(nova_linha)

    # Para cada exemplo na coluna EVA (alunos que evadiram)
    for _ in range(row['EVA']):
        # dados_expandidos.append({'AE': 'SIM', 'COM': row['COM'], 'EVA': row['EVA']})
        nova_linha = row.copy()  # Copiar todos os dados da linha atual
        nova_linha['AE'] = 'SIM'  # Adicionar a coluna AE com valor "SIM"
        dados_expandidos.append(nova_linha)
```

```python
# Criar o novo DataFrame a partir dos dados expandidos
df_exp = pd.DataFrame(dados_expandidos)
df_exp
```
